```python
import dense_correspondence_manipulation.utils.utils as utils
import os

utils.add_dense_correspondence_to_python_path()
utils.set_cuda_visible_devices([1])

from dense_correspondence.evaluation.evaluation import DenseCorrespondenceEvaluation, DenseCorrespondenceEvaluationPlotter
from dense_correspondence.dataset.spartan_dataset_masked import SpartanDataset
import logging
logging.basicConfig(level=logging.INFO)

import pandas as pd
import numpy as np
import time

DCEP = DenseCorrespondenceEvaluationPlotter
DCE = DenseCorrespondenceEvaluation
```

```python
def run_model(model_name="elephant_unit_8", dataset=None, save_folder_name="analysis", train_type="simulated"):
    model_folder = "/home/davidtseng/pytorch-dense-correspondence/data_volume/pdc/trained_models/{}/{}".format(train_type, model_name)
    DCE = DenseCorrespondenceEvaluation
    start_time = time.time()
    DCE.run_evaluation_on_network(model_folder, num_image_pairs=100, 
                                  cross_scene=False, compute_descriptor_statistics=True, 
                                  dataset=dataset, save_folder_name=save_folder_name)
    print "elapsed time = %d seconds" %(time.time() - start_time)
    
def get_dataset(yaml_name="simulated_elephant.yaml"):
    config_filename = os.path.join(utils.getDenseCorrespondenceSourceDir(), 'config', 'dense_correspondence', 
                               'dataset', 'composite', yaml_name)

    config = utils.getDictFromYamlFilename(config_filename)
    dataset = SpartanDataset(config=config)
    return dataset
```
